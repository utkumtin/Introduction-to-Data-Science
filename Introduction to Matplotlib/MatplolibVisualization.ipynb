```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyArray1 = np.linspace(0, 10, 20)
numpyArray2 = numpyArray1 ** 2
```

```python
(myFigure, myAxis) = plt.subplots()

myAxis.plot(numpyArray1, numpyArray2, color= "#BB26C4", alpha= .5)
myAxis.plot(numpyArray2, numpyArray1, "r")
```

```python
(newFigure, newAxis) = plt.subplots()
newAxis.plot(numpyArray1, numpyArray1 + 2, color= "blue", linewidth= 3)

newAxis.plot(numpyArray1, numpyArray1 + 3, color= "red", linewidth= 1.0)

newAxis.plot(numpyArray1, numpyArray1 + 4, color= "g", linestyle= "--")
newAxis.plot(numpyArray1, numpyArray1 + 5, color= "g", linestyle= ":")
newAxis.plot(numpyArray1, numpyArray1 + 6, color= "g", linestyle= "-.")

newAxis.plot(numpyArray1, numpyArray1 + 7, color= "#000000", linestyle= "--", marker= "o", markerfacecolor="r")
newAxis.plot(numpyArray1, numpyArray1 + 8, color= "#000000", linestyle= "--", marker= "o", markersize= 8, markerfacecolor="r")
```

```python
plt.scatter(numpyArray1, numpyArray2)
```
