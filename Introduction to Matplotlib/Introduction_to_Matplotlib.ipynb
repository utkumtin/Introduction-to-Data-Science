```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ageList = [10 , 20, 30, 30, 30, 40, 50, 60, 70, 75]
weightList = [20, 60, 80, 85, 86, 87, 70, 90, 95, 90]
```

```python
numpyAgeList = np.array(ageList)
numpyWeightList = np.array(weightList)
```

```python
plt.plot(numpyAgeList, numpyWeightList, "r")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Weight Change by Age")
# plt.show() if you did not use the %matplotlib inline
```

```python
numpyArray1 = np.linspace(0, 10, 20)
```

```python
numpyArray1
```

```python
numpyArray2 = numpyArray1 ** 3
```

```python
numpyArray2
```

```python
plt.plot(numpyArray1, numpyArray2, "r*-")
```

```python
plt.subplot(1,2,1) # 1 row, 2 columns, draw the first graph
plt.plot(numpyArray1, numpyArray2, "r*-")

plt.subplot(1,2,2) # 1 row, 2 columns, draw the second graph
plt.plot(numpyArray2, numpyArray1, "b--")
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.2, 0.2, 0.9, 0.9])
figureAxes.plot(numpyArray1, numpyArray2, "g")
figureAxes.set_xlabel("X Axis Data Name")
figureAxes.set_ylabel("Y Axis Data Name")
figureAxes.set_title("Graph Title")
```

```python
figure2 = plt.figure()

axis1 = figure2.add_axes([.1, .1, .9, .9])
axis2 = figure2.add_axes([.2, .5, .3, .3]) # loc on x axis, loc on y axis, width, height

axis1.plot(numpyArray1, numpyArray2, "r")
axis1.set_xlabel("X axis")
axis1.set_ylabel("Y axis")
axis1.set_title("Main Graph Title")

axis2.plot(numpyArray2, numpyArray1, "b")
axis2.set_xlabel("X axis Small")
axis2.set_ylabel("Y axis Small")
axis2.set_title("Small Graph Title")
```
